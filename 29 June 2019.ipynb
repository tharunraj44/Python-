```
#Drawing a square
import turtle as tt
a1=tt.Turtle()
a1.forward(150)
a1.right(90)
a1.forward(150)
a1.right(90)
a1.forward(150)
a1.right(90)
a1.forward(150)
a1.right(90)
tt.done()
```

```
#Drawing a square
import turtle as tt
a1=tt.Turtle()
a1.backward(150)
a1.left(90)
a1.backward(150)
a1.left(90)
a1.backward(150)
a1.left(90)
a1.backward(150)
a1.left(90)
tt.done()
```

```
#Drawing a square with  looping statements
import turtle as tt
a1=tt.Turtle()
for x in range(4):
    a1.forward(150)
    a1.right(90)
tt.done()
```

```
# spiralling star
import turtle as tt
a1=tt.Turtle()
a1.pencolor('red')
for x in range(80):
    a1.forward(x * 10)
    a1.right(144)
    
tt.done()
```

```
# spiralling star
import turtle as tt
a1=tt.Turtle()
a1.pencolor('red')
for x in range(250):
    a1.forward(200)
    a1.right(123)
a1.pencolor('green')
for x in range(250):
    a1.forward(200)
    a1.right(123)
    
    
tt.done()
```

```
# A colorful hexagon
colors=['red','orange','yellow','green','blue','indigo','violet','brown','lime','grey']
import turtle as tt
a1=tt.Turtle()
for x in range(360):
    a1.pencolor(colors[x%10])
    a1.width(x/100+1)
    a1.forward(x)
    a1.left(59)
tt.done()
```

```
import turtle as tt
a1=tt.Turtle()
color=['red','green','yellow','orange','blue','purple']
dot_distance=10
width=10
height=10
a1.penup()
for x in range(height):
    a1.pencolor(color[x%6])
    for i in range(width):
        a1.dot()
        a1.forward(dot_distance)
    a1.forward(dot_distance*width)
    a1.right(90)
    a1.forward(dot_distance)
    a1.left(90)
```

```
#generating blue flowers
from turtle import *
import random

for n in range(50):
    penup()
    goto(random.randint(-400,400),random.randint(-400,400))
    pendown()
    red_amount=random.randint(0,30)/100.0
    blue_amount=random.randint(50,100)/100.0
    green_amount=random.randint(0,30)/100.0
    pencolor((red_amount,blue_amount,green_amount))
    circle_size=random.randint(10,40)
    pensize(random.randint(4,6))
    for i in range(6):
        circle(circle_size)
        left(60)
tt.done()
```

```
# symbol generation using python
from turtle import *
def main():
    colors=["red","orange","yellow","green","blue","pink","purple"]
    reset()
    Screen()
    up()
    goto(-320,-195)
    width(70)
    for pcolor in colors:
        color(pcolor)
        down()
        forward(640)
        up()
        backward(640)
        left(90)
        forward(66)
        right(90)
    width(40)
    color("white")
    goto(0,-170)
    down()
    circle(170)
    left(90)
    forward(340)
    up()
    left(180)
    forward(170)
    right(45)
    down()
    forward(170)
    up()
    backward(170)
    left(90)
    down()
    forward(170)
    up()
    goto(0,300)
if __name__=="__main__":
    main()
    mainloop()
```

```
from turtle import *
N=80
def f(x):
    return 3.9*8*(1-x)
def g(x):
    return:3.9*(x-x**2)
def h(X):
    return 3.9*x-3.9*x*x
def jumpto(x,y):
    penup()
    goto(x,y)
def line(x1,y1,x2,y2):
    jumpto(x1,y1)
    pendown()
    jumpto(x2,y2)
def coosys():
    line(-1,0,N+1,0)
    line(0,-0-1,0,1.1)
def plot(fun,start,color):
    pencolor(color)
    x=start
    jumpto(0,x)
    pendown()
    dot(5)
    for i in range(N):
        x=fun(x)
        goto(i+1,x)
        dot(5)
def main():
    reset()
    setworldcoordinates(-1.0,-1.0,N+1,1.1)
    speed(0)
    hideturtle()
    coosys()
    plot(f,0.35,"blue")
    plot(g,)
```
