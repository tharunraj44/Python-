```
#Function to create a file and write some data
def createfile(filename):
    f=open(filename,"w")
    for i in range(10):
        f.write("This is %d Line\n" % i)
    print("file is created successfully and data is written")    
    f.close()
    return
createfile("file.txt")  
```

```
#function for reading the file data
def readfile(filename):
    f=open(filename,"r")
    if f.mode=="r":
        x=f.read()
        print(x)
    f.close()
    return
readfile("file.txt")  
```

```
#data to append
#function to append data to existing file
def appenddata(filename):
    f=open(filename,"a")
    f.write("New line1\n")
    f.write("New line2\n")
    return
appenddata("file1.txt")  
```

```
#Word count using file I/O python
def dataanalysiswordcount(filename,word):
    f=open(filename,"r")
    if f.mode=="r":
        x=f.read()
        lst=x.split()
    cnt=lst.count(word)
    return cnt
print(dataanalysiswordcount("file1.txt","Test"))  
    
    
```

```
#Function to count of characters in the file
def countcharacter(filename):
    f= open(filename,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
    return len(lst)
print(countcharacter("file1.txt"))  
```

```
#Function to count the upper case character from the given list
def countupper(filename):
  cntupper=0
  f=open(filename,"r")
  if f.mode=="r":
      x=f.read()
      lst=list(x)
  for i in lst:
      if i.isupper():
          cntupper+=1
  return cntupper
print(countupper("file1.txt"))  
```

```
#function  to count of lines in given list
def countlines(filename):
    f=open(filename,"r")
    if f.mode=="r":
        x=f.read()
        lst=x.split("\n")
    return len(lst)
countlines("file1.txt")
```

```
import re
def phonenumbervalidate(phone):
    pattern='^[6-9][0-9]{9}$'
    phone =str(phone)
    if re.match(pattern,phone):
        return True
    return False
print(phonenumbervalidate(996633855))
print(phonenumbervalidate(99668855))
```

```
#RE--Validation of Roll number
#152U1A0501

import re
def validaterollnumber(number):
    number=str(number)
    pattern="^[1][5][2][U][1][A][0][1-9][0-6][0-9]$";
    if re.match(pattern,number):
        return True
    return False
print(validaterollnumber("152U1A0555"))  
```

```
import re
def validateemailid(email):
    pattern="^[0-9a-z][0-9a-z.]{,13}[0-94a-z][@][a-z0-9]{3,18}[.][a-z]{2,4]$"
    if re.match(pattern,email):
        return True
    return False
print(validateemailid("anil.1889@gmail.com"))  
  
```

```
def validatepassword(s):
    pattern ="^[a-zA-z0-9!@#$]{6,15}$"
    if re.match(pattern,s):
        return True
    return False
print(validatepassword("Anil1889@21#M"))
print(validatepassword("Anil1889@21#M111122222"))
```
