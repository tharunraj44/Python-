##Bubble sort
it compares the adj items,if the first item is greater than second item then there would require of swap the items 
The complex-n*n

```
def bubblesort(a):
    for i in range (len(a)):
        for j in range (len(a)-1):
            if a[j]>a[j+1]:
               a[j],a[j+1]=a[j+1],a[j]
    for i in range (len(a)):
        print(a[i],end=" ")
list1=[19,1,25,6,18,3]
bubblesort(list1)
list1.sort()
print(list1)
            
```
